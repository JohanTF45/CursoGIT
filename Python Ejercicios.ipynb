````python
# 2. Escriba una función que reciba como parámetro un valor entero, el cual represente un número determinado de días.
#La función debe devolver el número de semanas y el número de días remanentes.
#Por ejemplo, para un valor de 10 días, el resultado debería verse así:

def semdias(dias):
    semana = 0
    for i in range(dias):
        if dias > 7:
            semana += 1
            dias -= 7
    return semana , dias
semana , dia = semdias(30)
print("{} semana(s) y {} día(s)".format(semana, dia))

def semdias2(dias):
    semanas = dias // 7
    dia = dias % 7
    return semanas , dia
semanas , dia = semdias2(30)
print("{} semana(s) y {} día(s)".format(semana, dia))
````

````python
#Escriba un bucle `while` para devolver el factorial de un numero entero, el factorial de 6 debe dar como resultado 720 (1x2x3x4x5x6)

num = 1
rango = 6
factorial = 1

# Complemente el código

while num <= rango:
    factorial *= num
    num += 1
print(factorial)
````

````python
#Escriba un bloque código usando bucles y condiciones para determinar,
#si un número dentro de la lista `es_primo` es número primo o no.

es_primo = [26, 39, 51, 53, 57, 79, 85]

# Complemente el código

for primo in es_primo:
    rango = 1
    while rango <= len(es_primo):
        rango += 1
        numero = primo%rango
        if numero == 0:
            print("{} NO es un número primo, porque {} es un factor de {}".format(primo, rango, primo ))
            break;
    else:
        print("{} ES un número primo".format(primo))
````

````python
# Escribe un ejercicio de clases con los puntos indicados

# Declaramos la clase TransporteAuto
class TransporteAuto():
# declaramos el metodo __init__
    def __init__(self):
        self.tipo=input("Tipo de Automovil : ")

# declaramos el metodo mostrar
    def mostrar(self):
        print("Tipo de automovil : ",self.tipo)
# declaramos la clase Automovil (hereda los atributos y metodos de la clase TransporteAuto)

class Automovil(TransporteAuto):
# declaramos el metodo __init__
    def __init__(self):
# llamamos al metodo init de la clase padre, utilizamos la funcion super() para hacer referencia al padre
        super().__init__()
        self.antiguedad=int(input("Año del Automovil : "))
# declaramos el metodo mostrar
    def mostrar(self):
        super().mostrar()
        print("El año del auto movil es  : ",self.antiguedad)
# declaramos el metodo es_antiguedad
    def antig(self):
        if self.antiguedad <1900:
            print("Su auto es antiguo")
        else:
            print("Su auto no es antiguo")

# comprobara si es o no antiguendad
# bloque principal - imprime la clase Automovil que hereda de TransporteAuto

Auto = Automovil()
Auto.mostrar()
Auto.antig()
````

````python
def planea_fiesta(bocadillos, personas):
    sobrantes = None
    por_persona = None
    try:
        por_persona = (bocadillos // personas)
        sobrantes = bocadillos % personas
    except ZeroDivisionError:
        print("Divicion entre 0")
    return(por_persona, sobrantes)

seguir_fiesta = 's'

while seguir_fiesta == 's':
    bocadillos = int(input("¿Cuántos bocadillos está preparando? "))
    personas = int(input("¿Cuántas personas van a asistir? "))
    por_persona, sobrantes = planea_fiesta(bocadillos, personas)

    if por_persona:  # si por_persona no es None
        mensaje = "\n¡Hagamos fiesta! Tendremos {} invitados, cada uno tendrá {} bocadillos y sobrarán {}."
        print(mensaje.format(personas, por_persona, sobrantes))
    seguir_fiesta = input("\n¿Desea continuar la fiesta? (s o n)")
````
